```python
class Rectangle:
  def __init__(self, width: int, height: int):
    self.width = width
    self.height = height

  def get_area(self):
    return self.width * self.height
```

```python
width = int(input("width = "))
```

```python
height = int(input("height = "))
```

```python
rec = Rectangle(width, height)
print(f"Area = {rec.get_area()}")
```

```python
class Game:
  def __init__(self, low=1, high=10) -> None:
    self.low = low
    self.high = high

  @staticmethod
  def guess_proc(low, high):
    number = (low + high) // 2
    answer = input(f"Загаданное число - {number} (y/n)?")
    if answer == "y":
      print("Угадал!")
    else:
      answer = input(f"Загаданное число меньше {number} (y/n)?")
      if answer == "y":
        Game.guess_proc(low, number-1)
      else:
        Game.guess_proc(number+1, high)

  def run_game(self):
    print("Загадайте число")
    Game.guess_proc(self.low, self.high)
```

```python
while True:
  answer = input("Хотите сыграть в игру (y/n)?")
  if answer == "y":
    game = Game()
    game.run_game()
  else:
    break
```
